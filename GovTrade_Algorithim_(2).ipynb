```python
import turicreate as turi;
import yfinance as fin;
import datetime as date;
```

```python
trade_data = turi.SFrame('all_transactions.csv'); 
trade_data
```

```python
tradeDataSimple = trade_data[trade_data['type']=='Sale (Full)']   
tradeDataSimple2 = trade_data[trade_data['type']=='Purchase']   
trade_data = tradeDataSimple.append(tradeDataSimple2)

trade_data
```

```python


# Step 2: Data Cleaning (optional)
# Depending on your analysis, you might want to clean the data by handling missing values or converting data types.

# Step 3: Filter Buy and Sell Entries
buy_entries = trade_data[trade_data['type'] == 'Purchase']
sell_entries = trade_data[trade_data['type'] == 'Sale (Full)']

# Step 4: Matching Buy and Sell Entries
matched_trades = []
for ticker in set(buy_entries['ticker']):
    buy_trades_ticker = buy_entries[buy_entries['ticker'] == ticker]
    sell_trades_ticker = sell_entries[sell_entries['ticker'] == ticker]
    
    # Check if there are corresponding buy and sell entries
    if len(buy_trades_ticker) > 0 and len(sell_trades_ticker) > 0:
        # Sort both buy and sell entries by the transaction_date to ensure they are in chronological order
        buy_trades_ticker = buy_trades_ticker.sort('transaction_date')
        sell_trades_ticker = sell_trades_ticker.sort('transaction_date')
        
        # Match buy and sell entries for this ticker
        for buy_row in buy_trades_ticker:
            # Find the sell trade with the same name as the buy trade
            matching_sell = sell_trades_ticker[(sell_trades_ticker['senator'] == buy_row['senator']) & (sell_trades_ticker['amount'] == buy_row['amount'])]

            if len(matching_sell) > 0:
                buy_date = buy_row['transaction_date']
                sell_date = matching_sell[0]['transaction_date']
                senator = buy_row['senator']
                amount = buy_row['amount']
                matched_trades.append((ticker, buy_date, sell_date,senator,amount))

# Step 5: Displaying Results
matched_sframe = turi.SFrame({'ticker': [trade[0] for trade in matched_trades],
                            'buy_date': [trade[1] for trade in matched_trades],
                            'sell_date': [trade[2] for trade in matched_trades],
                            'senator': [trade[3] for trade in matched_trades],
                            'amount': [trade[4] for trade in matched_trades]})
                            

```

```python
def convert_date_format(date_str):
    if(len(date_str)<6):
        date_str = date.today()
        return
    

    
    # Split the date string into month, day, and year components
    month, day, year = date_str.split('/')
    
    # Reformat the date string to "Year-Month-Day" format
    new_date_str = f"{year}-{month}-{day}"
    
    return new_date_str

matched_sframe['buy_date'] = matched_sframe['buy_date'].apply(convert_date_format)
matched_sframe['sell_date'] = matched_sframe['sell_date'].apply(convert_date_format)



def get_stock_price(ticker, date):
    stock_data = fin.download(ticker, start=date, end=date)
    return stock_data['Close'].iloc[1]

# Create a new column 'buy_price' and populate it with the stock prices when bought
#matched_sframe['buy_price'] = matched_sframe.apply(lambda row: get_stock_price(row['ticker'], row['buy_date']), dtype=float)

# Display the updated SFrame
print(matched_sframe)
```
